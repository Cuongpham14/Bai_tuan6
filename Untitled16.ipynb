```python
#EX2
# Step 1
import pandas as pd

# Step 2 and 3
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, delimiter='\t')

# Step 4
print(chipo.head(10))

# Step 5
print(len(chipo))

# Step 6
print(chipo.shape[1])

# Step 7
print(chipo.columns)

# Step 8
print(chipo.index)

# Step 9
chipo.groupby('item_name')['quantity'].agg(['count']).sort_values(by='count', ascending=False).head(1)

# Step 10
chipo.groupby('item_name')['quantity'].agg(['count']).sort_values(by='count', ascending=False).head(1)

# Step 11
chipo.groupby('choice_description')['quantity'].agg(['count']).sort_values(by='count', ascending=False).head(1)
# Step 12
chipo['quantity'].sum()

# Step 13.a
print(chipo['item_price'].dtype)


# Step 13.b
price_type = lambda x: float(x[1:-1])
print(price_type)
chipo.item_price = chipo.item_price.apply(price_type)

# Step 13.c
print(chipo['item_price'].dtype)

# Step 14
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print(revenue)

# Step 15
orders = chipo['order_id'].nunique()
print(orders)

# Step 16
average_revenue = revenue/orders
print(average_revenue)


# Step 17
items_sold = chipo['item_name'].nunique()
print(items_sold)
```
